 2) Seleccione a los empleados cuyo salario es mayor que 20.000 y menor que 50.000 euros (Colectivo 1)
   
   a) Sobre ese colectivo, ¿Cual es la frecuencia relativa acumulada de 30 días de vacaciones?


```r
GRUPO <- subset(datos,datos$salario>14000)
GRUPO <- subset(GRUPO,GRUPO$salario<34000)
summary(GRUPO)
```

```r
tabla.frecuencias = function(x)
{
  frec.n=table(x)
  #frec.n=table(X) proporciona la frecuencia absoluta que representa cada valor en la distribución X
  frec.N=cumsum(frec.n)
  #frec.N=cumsum(frec.n) proporciona la frecuencia absoluta acumulada
  frec.f= 100*prop.table(frec.n)
  #frec.f= 100*prop.table(frec.n) proporciona la frecuencia relativa
  frec.F = cumsum(frec.f)
  #frec.F = cumsum(frec.f) proporciona la frecuencia relativa acumulada
  cbind(frec.n,frec.N,frec.f,frec.F)
  #combinamos todo por columnas usando cbind()
}

tabla.frecuencias(GRUPO$vacaciones)
# Es 99.54%
```

 c) Dibuja el diagrama de tarta de la variable sexo

```r
TABLA_FRECUENCIAS <- table(GRUPO$sexo)
pie(TABLA_FRECUENCIAS, 
    labels=paste(names(TABLA_FRECUENCIAS), 
                 " ", 
                 TABLA_FRECUENCIAS, 
                 " ", 
                 round(TABLA_FRECUENCIAS* 100/sum(TABLA_FRECUENCIAS), 1), 
                 "%"),
    xlab="", ylab="",
    main="Distribución por sexo (Salarios entre 20 y 50 keuros)", 
    cex=0.9,
    cex.main=1.2 )
```

d) Dibuja el diagrama de barras apliadas de Categoría y Mercado

```r
TABLA_FRECUENCIAS2 <- table(GRUPO$categoria,GRUPO$mercado)
TABLA_FRECUENCIAS2

barplot(TABLA_FRECUENCIAS2,
        main="Distribucion dpor categorías y sexo", 
        xlab="Sexo",
        ylab="Frecuencias absolutas", 
        legend = rownames(TABLA_FRECUENCIAS2),
        col=c("red","blue","green") )
```
